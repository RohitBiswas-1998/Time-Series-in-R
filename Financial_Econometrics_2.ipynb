```
library(ggplot2)
```

```
data <- read.csv("Trade Openness.csv")
```

```
data
```

```
trade_open_raw <- data[, 5]

```

```
#Installing the Time Series package
install.packages("tseries")
library(tseries)
```

```
#Converting the raw data to time series data
trade_open_ts <- ts(trade_open_raw, start = 1960, frequency = 1)

class(trade_open_ts)
```

```
#Raw plot of the data (1980 - 2018)
plot(trade_open_ts, col = 2,lwd = 2, xlab = "Years", ylab = "Trade Openness", main = "Trade Openness in India")
abline(reg = lm(trade_open_ts~time(trade_open_ts)), col = 4, lwd = 2)
```

```
#Specifying the years used to fit the data using ARIMA
trade_open_data <- data[1:49, 5]
trade_open <- ts(trade_open_data, start = 1960, frequency = 1)
```

```
#Checking for stationarity of data
adf.test(trade_open)
```

```
#1st difference 
trade_open_diff1 <- diff(trade_open)
plot(trade_open_diff1, col = 4, lwd = 2, xlab = "Years", ylab = "First Difference", main = "Plot of 1st differenced data")
abline(reg = lm(trade_open_diff1 ~ time(trade_open_diff1)), col = 1, lwd = 2)
adf.test(trade_open_diff1)
```

```
#2nd difference
trade_open_diff2 <- diff(diff(trade_open))
plot(trade_open_diff2, col = 4, lwd = 2, xlab = "Years", ylab = "Second Difference", main = "Plot of 2nd differenced data")
abline(reg = lm(trade_open_diff2 ~ time(trade_open_diff2)), col = 1, lwd = 2)
adf.test(trade_open_diff2)
```

```
#Plotting the PACF to check for order of AR
pacf(trade_open_diff2, col = 3, lwd = 2, main = "PACF of 2nd Differenced data")
## significant spikes are obtained in lag 1 and 3
```

```
#Plotting the ACF to check for order of MA
acf(trade_open_diff2, col = 8, lwd = 2, main = "ACF of 2nd Differenced data")
## significant spikes are obtained in lags 0,1,3 and 4
```

```
#Installing the Forecast package
install.packages("forecast")
library(forecast)
```

```
#Un-doing the 2nd difference to obtain forecasts of "Trade Openness", and not the "Changes in Trade Openness"
trade_open_undiff <- diffinv(trade_open_diff2, xi = trade_open_diff1[1])
trade_data <- diffinv(trade_open_undiff, xi = trade_open[1])
```

```
#Specifying the various ARIMA models (order of 'p' and 'q' are determined from the PACF & ACF plots respectively)
model_1 <- arima(trade_data, order = c(0,2,1))
model_2 <- arima(trade_data, order = c(0,2,3))
model_3 <- arima(trade_data, order = c(0,2,4))
model_4 <- arima(trade_data, order = c(1,2,0))
model_5 <- arima(trade_data, order = c(1,2,1))
model_6 <- arima(trade_data, order = c(1,2,3))
model_7 <- arima(trade_data, order = c(1,2,4))
model_8 <- arima(trade_data, order = c(3,2,0))
model_9 <- arima(trade_data, order = c(3,2,1))
model_10 <- arima(trade_data, order = c(3,2,3))
model_11 <- arima(trade_data, order = c(3,2,4))
```

```
#AIC & BIC scores of all the models
scores_AIC <- AIC(model_1, model_2, model_3, model_4, model_5, model_6, model_7, model_8, model_9, model_10, model_11)
scores_BIC <- BIC(model_1, model_2, model_3, model_4, model_5, model_6, model_7, model_8, model_9, model_10, model_11)
```

```
aic <- scores_AIC$AIC
bic <- scores_BIC$BIC
```

```
#Comparing the AIC & BIC scores to select the best model
scores <- data.frame(aic, bic)
scores
```

```
#Residual plots
pacf(model_8$residuals, col = 3, lwd = 2, xlab = "Lag", ylab = "Partial Autocorrelations of error", main = "Model-8 PACF")
acf(model_8$residuals, col = 8, lwd = 2, xlab = "Lag", ylab = "Autocorrelations of error", main = "Model-8 ACF")
```

```
pacf(model_3$residuals, col = 3, lwd = 2, xlab = "Lag", ylab = "Partial Autocorrelations of error", main = "Model-3 PACF")
acf(model_3$residuals, col = 8, lwd = 2, xlab = "Lag", ylab = "Autocorrelations of error", main = "Model-3 ACF")
```

```
pacf(model_9$residuals, col = 3, lwd = 2, xlab = "Lag", ylab = "Partial Autocorrelations of error", main = "Model-9 PACF")
acf(model_9$residuals, col = 8, lwd = 2, xlab = "Lag", ylab = "Autocorrelations of error", main = "Model-9 ACF")
```

```
pacf(model_1$residuals, col = 3, lwd = 2, xlab = "Lag", ylab = "Partial Autocorrelations of error", main = "Model-1 PACF")
acf(model_1$residuals, col = 8, lwd = 2, xlab = "Lag", ylab = "Autocorrelations of error", main = "Model-1 ACF")
```

```
#Forecasting next 10 year values for 'Trade Openness in India'
model_8_pred <- forecast(model_8, h = 10, level = c(90, 95))

model_3_pred <- forecast(model_3, h = 10, level = c(90, 95))

model_9_pred <- forecast(model_9, h = 10, level = c(90, 95))

model_1_pred <- forecast(model_1, h = 10, level = c(90, 95))
```

```
#Calculating the Error in Forecasting
actuals <- trade_open_ts[50:59]
```

```
#Model_8
model_8_error <- ((actuals - model_8_pred$mean)/actuals)
```

```
#Model_3
model_3_error <- ((actuals - model_3_pred$mean)/actuals)
```

```
#Model_9
model_9_error <- ((actuals - model_9_pred$mean)/actuals)
```

```
#Model_1
model_1_error <- ((actuals - model_1_pred$mean)/actuals)
```

```
mean_error <- c(mean(model_8_error), mean(model_3_error), mean(model_9_error), mean(model_1_error))
mean_error
```

```
## Model_1 has the lowest error percentage 
```
